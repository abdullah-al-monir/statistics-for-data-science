```python
# binomial_distribution_plot.ipynb

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Parameters for the example
n = 10  # Number of trials (e.g., coin flips)
p = 0.5 # Probability of success (e.g., probability of heads)
x = np.arange(0, n + 1) # Possible number of successes (0 to 10)

# 1. Probability Mass Function (PMF)
pmf = binom.pmf(x, n, p)

plt.figure(figsize=(8, 6))
plt.bar(x, pmf, color='skyblue', edgecolor='black')
plt.title(f'Binomial Distribution PMF (n={n}, p={p})')
plt.xlabel('Number of successes (x)')
plt.ylabel('Probability')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 2. Cumulative Distribution Function (CDF)
cdf = binom.cdf(x, n, p)

plt.figure(figsize=(8, 6))
plt.plot(x, cdf, color='purple', marker='o', linestyle='-', linewidth=2)
plt.title(f'Binomial Distribution CDF (n={n}, p={p})')
plt.xlabel('Number of successes (x)')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.show()

# Example calculation: Probability of exactly 3 successes
probability_3_heads = binom.pmf(3, n, p)
print(f'Probability of exactly 3 successes in 10 trials: {probability_3_heads:.4f}')
# Output: Probability of exactly 3 heads in 10 trials: 0.1172
```
